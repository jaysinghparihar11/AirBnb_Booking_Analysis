## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
air_b_data_set = air_bnb_dataset.copy()
````

````python
air_b_data_set.drop(['last_review'], axis=1, inplace=True)
````

````python
air_b_data_set['host_name'].fillna('Unknown',inplace=True)
air_b_data_set['name'].fillna('blank',inplace=True)
````

````python
air_b_data_set["reviews_per_month"].fillna(0,inplace=True)
````

````python
Top5_neighbourhood = air_b_data_set.neighbourhood_group.value_counts().reset_index()
Top5_neighbourhood
````

````python
Neighbourhood_Groups_Reviews = air_b_data_set.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index().sort_values(by = 'number_of_reviews', ascending = False)
Neighbourhood_Groups_Reviews
````

````python
Minimum_Room_price = air_b_data_set.replace(0, np.nan, inplace=False).groupby(['neighbourhood_group','room_type'])['price'].min().unstack()
Minimum_Room_price
````

````python
Maximun_price_ = air_b_data_set.groupby('neighbourhood_group',as_index=False)['price'].max().sort_values(['price'],ascending = False).rename(columns = {'price':'Maximum price','neighbourhood_group':'Location'})
Maximun_price_
````

````python
most_listings_hosts = air_b_data_set.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()
most_listings_hosts.sort_values(by='calculated_host_listings_count',ascending=False)
````

````python
Min_availability = air_b_data_set.replace(0, np.nan, inplace=False).groupby(['neighbourhood_group'])['availability_365'].mean().sort_values(ascending=False)
Min_availability.round().reset_index()
````

### What all manipulations have you done and insights you found?

- I deleted the 'last_review' column from the dataset. Afterward, I resolved null value-related problems by using '.fillna'."
- Entire home/apt has the highest number of listings."
- I created columns 'most_listings_hosts', 'Min_availability', and 'Maximum_price_' for EDA.
